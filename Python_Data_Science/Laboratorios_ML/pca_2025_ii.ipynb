```python
import pandas as pd
columns = ['sepal-length','sepal-width','petal-length','petal-width','class']
iris = pd.read_csv('iris.data',names=columns)

X = iris[['sepal-length','sepal-width','petal-length','petal-width']]
X = X.values
```

```python
from sklearn.decomposition import PCA
n=4
pca = PCA(n_components=n)
pca.fit(X)
pca.explained_variance_ratio_
```

```python
import matplotlib.pyplot as plt
import numpy as np

componentes = list(range(1,n+1))
plt.figure(figsize=(15,6))
plt.plot(componentes, np.cumsum(pca.explained_variance_ratio_),'-o',label='components')
plt.legend()
plt.xlabel('components')
plt.ylabel('cumulative explained variance ratio')
plt.grid()
plt.title('pca')
plt.show()
```

```python
pca_model = PCA(n_components=2)
pca_model.fit(X)
pca_X = pca_model.transform(X)
pca_X
```

```python
y = iris[['class']]
y
```

```python
# Convert pca_X to a DataFrame
pca_df = pd.DataFrame(data=pca_X, columns=['pca1', 'pca2'])

# Reset index of y to ensure proper concatenation
y = y.reset_index(drop=True)

# Concatenate the pca_df and y DataFrames
final_df = pd.concat([pca_df, y], axis=1)

# Display the new DataFrame
final_df
```
