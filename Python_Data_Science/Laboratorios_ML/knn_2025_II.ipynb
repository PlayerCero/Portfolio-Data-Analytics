```python
import pandas as pd
columns = ['sepal-length','sepal-width','petal-length','petal-width','class']
iris = pd.read_csv('iris.data',names=columns)
iris.head()
```

```python
X = iris[['sepal-length','sepal-width','petal-length','petal-width']]
X = X.values
X
```

```python
y = iris[['class']]
y = y.values.ravel()
y
```

```python
X = (X - X.min(axis=0))/(X.max(axis=0) - X.min(axis=0))
X
```

```python
import numpy as np
import matplotlib.pyplot as plt
import sklearn.metrics as metrics
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
```

```python
nn = 20 # valores para k desde 1 hasta nn
n = 100 # total de aleatorizaciones
k = []
k_training_acc = []
k_testing_acc = []

for i in range(1,nn+1):
  k.append(i)
  training_acc = np.zeros(n)
  testing_acc = np.zeros(n)
  print("NN",i)
  for j in range(n):
    X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2,random_state=j)
    knn_classifier = KNeighborsClassifier(n_neighbors=i,algorithm='brute')
    knn_classifier.fit(X_train, y_train)

    y_train_hat = knn_classifier.predict(X_train)
    y_test_hat = knn_classifier.predict(X_test)

    training_acc[j] = metrics.accuracy_score(y_train, y_train_hat)
    testing_acc[j] = metrics.accuracy_score(y_test, y_test_hat)

k_training_acc.append(training_acc.mean())
k_testing_acc.append(testing_acc.mean())

```

```python
plt.figure(figsize=(15,8))
plt.plot(k,k_training_acc, '-o', label='training')
plt.plot(k,k_testing_acc,'-o', label='testing')
plt.legend()
plt.xlabel('k')
plt.ylabel('accuracy')
plt.grid()
plt.title('k tuning')
plt.show()
```

```python

X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2,random_state=32)
knn_classifier = KNeighborsClassifier(n_neighbors=6,algorithm='brute')
knn_classifier.fit(X_train, y_train)

y_train_hat = knn_classifier.predict(X_train)
y_test_hat = knn_classifier.predict(X_test)

print(metrics.accuracy_score(y_test,y_test_hat))
print(metrics.confusion_matrix(y_test,y_test_hat))
print(metrics.f1_score(y_test, y_test_hat, average=None))
print(metrics.classification_report(y_test, y_test_hat))
```
