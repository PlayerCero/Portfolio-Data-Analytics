```python
import numpy as np
import tensorflow as tf
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.metrics import classification_report
```

```python
names = ['sepal-length', 'sepal-width', 'petal-length', 'petal-width', 'class']
iris = pd.read_csv("iris.data",names=names)
#iris.head()
X = iris[["sepal-length","sepal-width","petal-length","petal-width"]]
y = iris[["class"]].values.ravel()
```

```python

# Codificar las etiquetas (de texto a números)
encoder = LabelEncoder()
labels_encoded = encoder.fit_transform(y)
labels_encoded
```

```python
# Normalizar las características
scaler = StandardScaler()
data_scaled = scaler.fit_transform(X)
data_scaled
```

```python

# Dividir en conjuntos de entrenamiento y prueba
X_train, X_test, y_train, y_test = train_test_split(data_scaled, labels_encoded, test_size=0.2)

# Crear la red neuronal
model = tf.keras.Sequential([

    # 1ra capa: aprender combinaciones no lineales de las 4 variables de entrada.
    tf.keras.layers.Dense(10, activation='relu', input_shape=(4,)),  # Capa oculta con 10 neuronas y capa de entrada implícita de 4 features

    # 2da capa: Reduce la dimensionalidad y combina las representaciones aprendidas por la primera capa.
    tf.keras.layers.Dense(8, activation='relu'),                    # Otra capa oculta

    # softmax convierte los valores en probabilidades que suman 1.
    tf.keras.layers.Dense(3, activation='softmax')                  # Capa de salida con 3 clases
])
# Compilar el modelo
model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])
```

```python
# Entrenar el modelo
history = model.fit(X_train, y_train, epochs=50, batch_size=10, validation_split=0.2)
```

```python
# Evaluar el modelo en el conjunto de prueba
test_loss, test_accuracy = model.evaluate(X_test, y_test, verbose=0)
print('Accuracy:',test_accuracy)
y_hat = model.predict(X_test)
y_hat
```

```python
y_hat = np.argmax(model.predict(X_test),axis=1)
y_hat
```

```python
# Generar un reporte de clasificación
y_pred = np.argmax(model.predict(X_test), axis=1)
print("\nReporte de clasificación:")
print(classification_report(y_test, y_pred, target_names=encoder.classes_))
```
